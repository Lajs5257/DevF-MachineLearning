# Introducción

## __¿Qué es el aprendizaje de maquina *(<font color='CD5C5C'>Machine learning</font>)*?__

*Machine learning* (<font color='CD5C5C'>ML</font>) es la ciencia (y el arte) de programar computadoras para que puedan aprender a partir de datos

\[ML es el\] campo de estudio que da a las computadoras la habilidad de aprender sin que sean explicitamente programadas.

-Arthur Samuel, 1959

Se dice que un programa computacional aprende de la experiencia **E** con respecto a alguna tarea **T** y una métrica de desempeño *(<font color='CD5C5C'>performance</font>)* **P**, si su *performance* en **T**, medido con **P**, mejora con la experiencia de **E**.

-Tom Mitchell, 1997

## __Tipos de ML__

Podemos clasificar de forma muy general los tipos de ML con base en los siguientes criterios:

* Si están entrenaddos (o no) con supervisión humana 
  * Aprendizaje supervisado (<font color='CD5C5C'>Supervised learning</font>)
  * Aprendizaje no supervisado (<font color='CD5C5C'>Unsupervised learning</font>)
  * Aprendizaje semi-supervisado (<font color='CD5C5C'>Semisupervised learning</font>)
  * Aprendizaje por refuerzo (<font color='CD5C5C'>Reinforcement learning</font>)
 
* Si pueden aprender o no de forma incremental sobre la marcha 
  * Aprendizaje en linea (<font color='CD5C5C'>Online learning</font>)
  * Aprendizaje por lotes (<font color='CD5C5C'>Batch learning</font>)

* Si trabajan simplemente comparando nuevos puntos de datos con puntos de datos conocidos, o en su lugar, detectando patrones en los datos de entrenamiento y construyendo un modelo predictivo
  * Aprendizaje basado en instancias (<font color='CD5C5C'>Instance-based learning</font>)
  * Aprendizaje basado en modelos (<font color='CD5C5C'>Model-based learning</font>)

Estos criterios no son exclusivos, pueden ser combinados!

### Supervised learning

En el aprendizaje supervisado, el conjunto de entrenamiento con el que alimentamos los algoritmos contiene ya las soluciones buscadas, estas soluciones son llamadas etiquetas (<font color='CD5C5C'>labels</font>)

churn.png

* Clasificación - El detector de spam, es entrenado con muchos emails junto con su clase (spam o verdadero), el algoritmo debe aprendes a clasificar casos nuevos.

* Regresión - pronosticar el precio de un auto, usando información como, marca, modelo, tipo de motor, kilometraje, etc. a esta información (variables) se le llama predictores. Para entrenar el algoritmo, es necesario que se den muchos ejemplos de autos que incluyan variables predictorias y precios (labels)

__NOTAS:__ 

* Regresión es un termino estadístico introducido por Francis Galton cuando estudiaba el hecho de que los hijos de personas altas tienden a ser mas bajos que sus padres. Como los hijos eran mas bajos, lo llamo regresión a la media. Este nombre fue aplicado a los métodos que uso para analizar correlaciones entre variables

* En ML un atributo *(<font color='CD5C5C'>atribute</font>)* es una variable (como el kilometraje), mientras que una característica *(<font color='CD5C5C'>feature</font>)* su significado dependerá del contexto, pero generalmente un feature significa un atributo junto con su valor (kilometraje = 10,000). En muchos textos, atribute y feature se usan de manera indiscriminada

Algunos de los algoritmos de aprendizaje supervisados mas importantes son:

* K vecinos cercanos *(<font color='CD5C5C'>k-Nearest Neighbors "KNN"</font>)*
* Regresión lineal *(<font color='CD5C5C'>Linear Regression</font>)*
* Regresión logística *(<font color='CD5C5C'>Logistic Regression</font>)*
* Maquinas de soporte vectorial *(<font color='CD5C5C'>Support Vector Machines "SVMs"</font>)*
* Árbol de desición y bosques aleatorios *(<font color='CD5C5C'>Decision Trees & Random Forests</font>)*
* Redes neuronales *(<font color='CD5C5C'>Neural networks</font>)*

__NOTA:__ Algunas arquitecturas de redes neuronales pueden ser no supervisadas como los *autoencoders* y las *restricted Boltzmann machines*. También pueden ser semisupervisadas como las *deep belief networks*.

### Unsupervised learning

unsupervised.png

La característica particular del aprendizaje no supervisado es que: "los datos de entrenamiento no cuentan con __*labels*__".

Algunos de los algoritmos de aprendizaje no supervisado mas importantes son:

#### Algoritmos de agrupamiento *__(<font color='CD5C5C'>Clustering</font>)__*

Son usados para crear grupos con características similares

  * K-Means
  * DBSCAN
  * Hierarchical Cluster Analysis (HCA)

#### Reducción de dimensiones *__(<font color='CD5C5C'>Dimensionality reduction</font>)__*

Los algoritmos de visualización son usados para entender mejor la estructura de los datos, estos regresan objetos en 2D o 3D con las representaciones de los datos originales tratando de preservar lo mejor posible sus estructuras internas a modo que no se traslapen.

Los algoritmos de reducción de dimensiones tienen como objetivo simplificar los datos sin perder mucha información. 

Por ejemplo el kilometraje de un auto esta fuertemente correlacionado con su edad, entonces el algoritmo de reducción de dimensiones juntara los dos features en uno que represente edad y kilometraje (Esto es llamado *Feature extraction*)

__TIP__ It is often a good idea to try to reduce the dimension of your training data using a dimensionality reduction algorithm before you feed it to another Machine Learning algorithm (such as a supervised learning algorithm). It will run much faster, the data will take up less disk and memory space, and in some cases it may also perform better.

  * Principal Component Analysis (PCA)
  * t-Distributed Stochastic Neighbor Embedding (t-SNE)
